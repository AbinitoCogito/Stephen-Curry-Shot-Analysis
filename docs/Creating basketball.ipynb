```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

Creating basketball with matplolib.pyplot. The module named patches and the modul is useful for plotting geometric shapes.

```python
fig, ax = plt.subplots()
basketball = patches.Circle((0.5, 0.5), radius=0.4, edgecolor='brown', facecolor='saddlebrown', lw=2)
ax.add_patch(basketball)

height_lines = patches.Rectangle((0.5,0.1), width=0.01, height=basketball.get_height(), edgecolor='black',facecolor='black')
ax.add_patch(height_lines)

width_lines = patches.Rectangle((0.1,0.5), width=basketball.get_width(), height=0.01, edgecolor='black',facecolor='black')
ax.add_patch(width_lines)

diagonal_line_1 = patches.ConnectionPatch((0.22, 0.22), (0.78, 0.78), 'data', 'data', color='black', lw=2)
diagonal_line_2 = patches.ConnectionPatch((0.78, 0.22), (0.22, 0.78), 'data', 'data', color='black', lw=2)

ax.add_patch(diagonal_line_1)
ax.add_patch(diagonal_line_2)

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')
plt.show()
```
