Investigating Curry's overall shot accuracy with Histogram Plot.

A histogram is a chart that plots the distribution of a numeric variable's values as a series of bars. Each bar typically covers a range of numeric values called a bin or class; a bar's height indicates the frequency of data points with a value within the corresponding bin.

```python
missmade = Curry['EVENT_TYPE']
plt.hist(missmade, edgecolor='black', alpha=0.75)
plt.title('Stephen Curry Field Goal Made/Miss Histogram Graphic ', fontsize=16)
plt.xlabel('Field Goal Missed/Made', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.text(0.5, 400, 'Made/Miss shots differentiation', fontsize=15, color='red', ha='center')
plt.axhline(y=655, color='red', linestyle='--', linewidth=2, label='Horizontal Line')
plt.show()
```

At first look, the graph could be interpreted wrong but this is actually impressive statistics about overall shooting history of Curry. Here is why:

```python
fgm_per = Curry['SHOT_MADE_FLAG'].sum() / Curry['SHOT_MADE_FLAG'].size
print(f'Stephen Curry field goal made percentage for 2014-2024: {fgm_per}')
```
